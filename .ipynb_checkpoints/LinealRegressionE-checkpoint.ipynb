## Regresión Multilineal y Métricas Implementadas

Utilizamos el modelo de regresión lineal para predecir el precio del jugador en 2016 con base en nuestro DataFrame X:

$ y_{pred} = \beta_{0} + \beta_{1}x_{1}+\beta_{2}x_{2}+\beta_{3}x_{3}+...+\beta_{n}x_{n}$

El procedimiento al implementar un modelo de Machine Learning es el siguiente:

* Se parten los datos disponibles en tres conjuntos, a saber: Entrenamiento ($ X_{train}$, $y_{train}$), Prueba ($X_{test}$, $y_{test}$) y Validación ($X_{valid}$, $y_{valid}$).

* El conjunto de Entrenamiento se usa para 'entrenar' al modelo, en el caso de la regresión lineal múltiple, esto se traduce a usar los datos para obtener los valores de $ \beta_{i}$.

* Una vez entrenado, se usa el conjunto prueba para determinar los metaparámetros óptimos para el modelo (no aplica para regresión lineal).

* Se usa el modelo entrenado con los metaparámetros correspondientes para predecir los valores de la variable objetivo en el conjunto de validación $X_{valid}$ obteniendo $y_{pred}$.

* Por último se compara la predicción $y_{pred}$ con el valor real $y_{valid}$ por medio de métricas que evalúan el rendimiento del modelo.

Las metricas de evaluación de rendimiento implementadas fueron las siguientes:

* Raíz del error cuadrático medio: 

$ RMSE = \sqrt{\frac{\Sigma (Y_{real}-Y_{pred})^{2}}{n}}  $

* Residuo promedio:

$ Residuo = \frac{1}{n} \Sigma \frac{|Y_{real}-Y_{pred}|}{Y_{real}} $

* R cuadrada:

$ R^{2} = \frac{\sigma^{2}_{XY}}{\sigma^{2}_{X} \sigma^{2}_{Y}}$


```python
from sklearn.metrics import mean_squared_error, r2_score, explained_variance_score
from sklearn.metrics import mean_absolute_error, median_absolute_error
```

Primero definimos la función para evaluar el rendimiento del modelo:

```python
def ev(y_pred,y_test):
    
    plt.figure(figsize = (10,6))
    res = abs(y_test-y_pred)/y_test * 100
    res.hist(bins = 100)
    plt.xlabel('Residuo', fontsize = 15)
    plt.ylabel('Cuenta de Residuos', fontsize = 15)
    plt.title('Análisis de Residuos', fontsize = 15)
    plt.xticks(fontsize = 12)
    plt.yticks(fontsize = 12)
    plt.xlim(res.quantile(0.005), res.quantile(0.99))
    plt.show()
        
    plt.figure(figsize = (10,6))
    plt.scatter(y_pred,y_test)
    plt.xlabel('Prediction', fontsize = 15)
    plt.ylabel('Real Value', fontsize = 15)
    plt.title('Predictions Scaterring', fontsize = 15)
    plt.xticks(fontsize = 12)
    plt.yticks(fontsize = 12)
    plt.show()
    
    print('Residuos en promedio son: %1.1f' % res.mean()+ ' por ciento')
    print('R^2: %1.2f' %r2_score(y_test,y_pred))
    print('Explained Variance %1.2f' %explained_variance_score(y_test, y_pred))
    print('Mean Absolute Error %1.2f' %mean_absolute_error(y_test, y_pred))
    print('Median Absolute Error %1.2f' %median_absolute_error(y_test,y_pred))
    print('Mean Squared Error: %1.2f' % np.sqrt(mean_squared_error(y_pred,y_test)))
```

## Modelos Implementados

Objetivo: probar el ranking como una variable predictiva para el valor del jugador en 2016

Con el presente dataset se realizaron los siguientes cuatro modelos de regresión lineal multiple:

* Primer Modelo: no se consideraron las variables ranking ni ranking restringido.

* Segundo Modelo: no se uso la variable ranking restringido.

* Tercer Modelo: no se uso la bariable ranking.

* Cuarto Modelo: se usaron todas las variables, en particular, el ranking restringido y el ranking.


### Primer Modelo: no tomar en cuenta ni el ranking ni el ranking restringido

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X1 = X.drop(['ranking_pos1', 'ranking_pos2', 'ranking_pos3', 'ranking_restringido_pos1', 'ranking_restringido_pos2',
       'ranking_restringido_pos3'], 1)
```

```python
sc_X = StandardScaler()
X_ = sc_X.fit_transform(X1)
```

```python
X_train_test, X_valid, y_train_test, y_valid = train_test_split(X_,
                                                                y, 
                                                                test_size = 0.1,
                                                                random_state = 8)

X_train, X_test, y_train, y_test = train_test_split(X_train_test,
                                                    y_train_test,
                                                    test_size = 0.1, 
                                                    random_state = 8)
```

```python
regressor = LinearRegression()

regressor.fit(X_train,y_train)

y_pred = regressor.predict(X_valid)

ev(y_pred,y_valid)
```

Validación:

```python
coef = regressor.coef_
coefs = pd.DataFrame(index = X1.columns, data = coef, columns = ['coeficientes'])
coefs = coefs.sort_values(by= 'coeficientes')

plt.figure(figsize = (10,10))
coefs.plot.barh(figsize=(10,10))
plt.title('Coeficientes de Regresión Lineal Múltiple', fontsize = 15)
plt.xlabel('Valor Numérico del Coeficiente', fontsize = 15)
plt.ylabel('Variable', fontsize = 15)
plt.xticks(fontsize = 10)
plt.yticks(fontsize = 10);
```
